# Q1

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url).text
toronto = BeautifulSoup(source, 'lxml')
```

```python
from IPython.display import display_html
tab = str(toronto.table)
display_html(tab,raw=True)
```

```python
df_all = pd.read_html(tab)
df=df_all[0]
df.head()
```

```python
indexNum = df[df['Borough'] == 'Not assigned'].index
df.drop(indexNum, inplace = True)
df.head(20)
```

```python
df_group = df.groupby(['Postal code', 'Borough'], sort = False).agg( ','.join)
df_fin = df_group.reset_index()
df_fin.head(20)
```

# Q3

### Getting all rows from df which contains Toronto in Borough

```python
toronto = df_mrg[df_mrg['Borough'].str.contains('Toronto',regex=False)]
toronto.head(20)
```

### Using KMeans for clustering

```python
k=5
clustering = toronto.drop(['Postal code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(clustering)
kmeans.labels_
toronto.insert(0, 'Clusters', kmeans.labels_)
```

```python
toronto.head()
```

### Visualizing the Clusters

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=11)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, neighborhood, cluster in zip(toronto['Latitude'], toronto['Longitude'], toronto['Neighborhood'], toronto['Clusters']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
